# Unsupervised Learning
K-means Clustering using custom data

```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
data=("E:/datasets-master/titanic.csv")
df=pd.read_csv(data)
df

```

```python
df=df.dropna()
```

```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt

x=df["Pclass"]
y=df["Age"]
data=list(zip(x,y))
```

```python
from sklearn.cluster import KMeans
model= KMeans(n_clusters=2)
model.fit(data)
```

```python
plt.scatter(x,y,c=model.labels_)
plt.show()
```
